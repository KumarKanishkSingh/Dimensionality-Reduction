```python
s_int = list(range(0,58))
FOLDERS = [str(x) for x in s_int]
```

```python
print(FOLDERS)
```

```python
imgdata=[]
```

```python
labels = []
```

```python
for i in s_int:
    folder=FOLDERS[i]
    path = os.path.join(DATADIR, folder) 
    for img in os.listdir(path):
            img_array = cv2.imread(os.path.join(path,img))
            new_array = cv2.resize(img_array, (100, 100))
            new_array.resize(30000)
            labels.append(i)
            imgdata.append(new_array)
```
